```python
import pandas as pd

df = pd.read_csv('../datasets/origin/categorize.csv')
df['is_product'] = df['category'].apply(lambda x: 'PRODUCT' in x)
product_voc = list(df.query('is_product == True')['mail'])
product_voc[:5]
```

比較

BERT

```python
from transformers import BertJapaneseTokenizer, BertForTokenClassification
from transformers import pipeline

model_name = 'jurabi/bert-ner-japanese'
tokenizer_name = 'jurabi/bert-ner-japanese'
# 'nlp-waseda/roberta-base-japanese'

model = BertForTokenClassification.from_pretrained(model_name)
tokenizer = BertJapaneseTokenizer.from_pretrained(tokenizer_name)

ner = pipeline('ner', model=model, tokenizer=tokenizer)

for text in product_voc[:5]:
    result = ner(text)
    print(text)
    print(result)
```

Luke

```python
from transformers import MLukeTokenizer,pipeline, LukeForTokenClassification

tokenizer = MLukeTokenizer.from_pretrained('Mizuiro-sakura/luke-japanese-base-finetuned-ner')
model=LukeForTokenClassification.from_pretrained('Mizuiro-sakura/luke-japanese-base-finetuned-ner') # 学習済みモデルの読み込み
ner = pipeline('ner', model=model, tokenizer=tokenizer)

for text in product_voc[:5]:
    result = ner(text)
    print(text)
    print(result)
```
